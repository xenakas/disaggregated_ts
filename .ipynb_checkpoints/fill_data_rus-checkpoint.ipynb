```r
library(ggplot2)
library(dplyr)
library(imputeTS)
library(data.table)
library(forecast)
```

```r
df_rus = read.csv(file="krim_nash.csv", header=TRUE, sep=",")

df_rus = as.data.table(df_rus)
```

```r
rownames(df_rus) = as.Date(df_rus$X0)
#rownames(data)
df_rus$X0 <- NULL
#colnames(data)
```

```r
df_rus$СЕВАСТРА[1:108] 
```

```r
model <- Arima(ts(rnorm(200),freq=4), order=c(2,0,0), seasonal=c(2,0,0))
foo <- simulate(model, nsim=108)
fit <- Arima(foo, order=c(2,0,0), seasonal=c(2,0,0))
```

```r
fit$fitted + 21
```

```r
df_rus$СЕВАСТРА[1:108]  <- fit$fitted * 20 - 20 
```

```r
ts.plot(df_rus$СЕВАСТРА) 
```
