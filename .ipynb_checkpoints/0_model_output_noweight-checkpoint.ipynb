```r
modelz  <-  c('model1_ar(2)_lin',
'model2_ar(12)_lin',
'model3_ar(12)_poly',
'model4_ari',
'model5_arma',
'model6_arima',
'model7_auto.arima',
'model8_ets',
'model9_auto.ets',
'model_10_tbats')
```

```r
a  <- b <-  list()
for (i in 1:length(modelz)){
    a[[i]]  <- readRDS(paste(modelz[i],'df_mean_acc_noweights.rds', sep='/'))
    b[[i]]  <- readRDS(paste(modelz[i],'plots_test_not_all_noweights.rds', sep='/'))
}

paramz = readRDS("df_params.rds")
```

```r
params_score  <-  a[[1]][1,]
for (i in 2:length(a)){
    params_score <-  rbind(params_score, a[[i]][1,])
}
```

```r
colnames(params_score)  <-  c('Квартальные', 'Сезонно.сглаженные', 'Месячные')
rownames(params_score)  <- c('AR_1 (лин.тренд)',  'AR_2 (лин.тренд)', 'AR_2 (квадр. тренд)', 'ARI', 'ARMA (лин.тренд)', 'ARIMA', 'ARIMA (автом. подбор)', 'ETS', 'ETS (автом. подбор)', 'TBATS' )
```

```r
params_score = params_score[order(params_score$Квартальные),]
params_score['order_eu'] = seq(1, dim(params_score)[1])
params_score = params_score[order(params_score$Сезонно.сглаженные),]
params_score['order_us'] = seq(1, dim(params_score)[1])
params_score = params_score[order(params_score$Месячные),]
params_score['order_rus'] = seq(1, dim(params_score)[1])
params_score = params_score[c('AR_1 (лин.тренд)',  'AR_2 (лин.тренд)', 'AR_2 (квадр. тренд)', 'ARI', 'ARMA (лин.тренд)', 'ARIMA', 'ARIMA (автом. подбор)', 'ETS', 'ETS (автом. подбор)', 'TBATS' ),]
```

```r
params_score['names_eu']  <- rownames(params_score)[params_score[,4]]
params_score['names_us']  <- rownames(params_score)[params_score[,5]]
params_score['names_rus']  <- rownames(params_score)[params_score[,6]]
```

```r
us_df  <-  t(us_df)
colnames(us_df)  <-  rownames(a[[1]])
rownames(us_df)  <-  c('model1_ar(2)_lin',
'model2_ar(12)_lin',
'model3_ar(12)_poly',
'model4_ari',
'model5_arma',
'model6_arima',
'model7_auto.arima',
'model8_ets',
'model9_auto.ets',
'model_10_tbats')
us_df = as.data.frame(us_df)
us_df$'1 level'  <-  NULL
```

```r
rus_df  <-  t(rus_df)
colnames(rus_df)  <-  rownames(a[[1]])
rownames(rus_df)  <-  c('model1_ar(2)_lin',
'model2_ar(12)_lin',
'model3_ar(12)_poly',
'model4_ari',
'model5_arma',
'model6_arima',
'model7_auto.arima',
'model8_ets',
'model9_auto.ets',
'model_10_tbats')
rus_df = as.data.frame(rus_df)
rus_df$'1 level'  <-  NULL
```
