```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('krim_nash.csv', index_col=0, header=[0], parse_dates=True)
```

```python
df = np.round(df.resample('M').ffill()/12)
```

```python
df = df.iloc[:-1,:]
```

```python
df.to_csv('krim_nash.csv')
```

```python
df2 = pd.read_csv('hts_rus_death_all.csv', index_col=0, header=[0], parse_dates=True)
```

```python
df2.loc[:,[i.startswith('КРЫМУ') for i in list(df2.columns)]]
```

```python
df3 = df2.loc[:,[i.startswith('КРЫМУ') for i in list(df2.columns)]].div(df2.loc[:,[i.startswith('КРЫМУ') for i in list(df2.columns)]].sum(axis=1), axis=0)[108:120]
df4 = df2.loc[:,[i.startswith('СЕВАУ') for i in list(df2.columns)]].div(df2.loc[:,[i.startswith('СЕВАУ') for i in list(df2.columns)]].sum(axis=1), axis=0)[108:120]
```

```python
df3.to_csv('123123.csv')
df4.to_csv('12312313.csv')
```

```python
df5 = pd.read_csv('123123.csv', header=[0], parse_dates=True)
```

```python
df5.index = df.index
```

```python
a = df5.loc[:,[i.startswith('КРЫМУ') for i in list(df5.columns)]]
b = df5.loc[:,[i.startswith('СЕВАУ') for i in list(df5.columns)]]
for i in range(a.shape[0]):
    a.iloc[i,:] = a.iloc[i,:] * df.iloc[:,1][i]
    b.iloc[i,:] = b.iloc[i,:] * df.iloc[:,3][i]
```

```python
a = round(a)
b = round(b)
```
