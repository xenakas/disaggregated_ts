```r
for (i in 1:4){
    if (i == 1){
        var_sort = mean
        var_group = NULL
    }else{var_sort = NULL
         if (i == 2){
             var_group = substr(colnames(df_eu), 1,2)
         }
          else if (i == 3){
             var_group = substr(colnames(df_eu), 3,3)
         }else{
               var_group = read.csv('index_list_eu.csv')[,2]
          }
         }
    
    k  <-  i + 0      
        
    list_plots[[k]]  <-  mvtsplot(df_eu, group = var_group,  norm = "internal", xtime = time(gdp_agg), sort = var_sort , level = 9, rowstat = mean )
}
```
