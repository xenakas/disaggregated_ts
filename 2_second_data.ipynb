### EC

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_gdp = pd.read_csv('hts_eu_a10.csv', index_col = 0, header = 0) #  header=[0,1], )
```

```python
# pd.MultiIndex.from_arrays(arrays, names=('number', 'color'))
```

```python
a = [i[0:2] for i in df_gdp.columns.values]
b = [i[2] for i in df_gdp.columns.values]
arrays = [a, b]
```

```python
df_gdp.columns = pd.MultiIndex.from_arrays(arrays)
```

```python
plt.plot(df_gdp.sum(level=0, axis=1));
```

```python
df_gdp.sum(level=0, axis=1).to_csv("hts_eu_a10_state.csv")
```

```python
df_gdp.columns = df_gdp.columns.swaplevel(0, 1)
```

```python
df_gdp.sort_index(axis=1, level=0, inplace=True)
```

```python
plt.plot(df_gdp.sum(level=0, axis=1));
```

```python
df_gdp.sum(level=0, axis=1).to_csv("hts_eu_a10_type.csv")
```

```python
df_ind = pd.read_csv('eu_index_list.csv', index_col = 0, header = 0) #  header=[0,1], )
```

```python
letters = 'abcdefghijklmnopqrstuvwxyz'
```

```python
np.unique(df_ind.iloc[:,1])[-1]
```

```python
keys = np.unique(df_ind.iloc[:,1])
values = [letters[i] for i in range(np.unique(df_ind.iloc[:,1])[-1])]
```

```python
keys
```

```python
dictionary = dict(zip(keys, values))
```

```python
df_ind.iloc[:,1] = df_ind.iloc[:,1].map(dictionary) 
```

```python
dictionary = dict(zip(df_ind.index, df_ind.iloc[:,1] ))
```

```python
dictionary
```

```python
df_gdp = pd.read_csv('hts_eu_a10.csv', index_col = 0, header = 0) #  header=[0,1], )
df_gdp.columns.map(dictionary)
```

```python
a = df_gdp.columns.map(dictionary)
b = df_gdp.columns.values
arrays = [a, b]

df_gdp.columns = pd.MultiIndex.from_arrays(arrays)

plt.plot(df_gdp.sum(level=0, axis=1));
```

```python
df_gdp.sum(level=0, axis=1).to_csv("hts_eu_a10_clust.csv")
```
