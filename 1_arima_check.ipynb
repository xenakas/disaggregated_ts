
## Aggregated TS

```r
df_eu  <-  read.csv(file="hts_eu_a10.csv", header=TRUE, sep=",", row.names =1 )
```

```r
gdp_agg <- df_eu_sums <-  rowSums(df_eu)
gdp_agg  <- ts(gdp_agg, frequency = 4, start = 2000) 
```

```r
train <- window(gdp_agg, end=c(2016,4))
test <- window(gdp_agg, start=c(2017,1))
```

```r
model  <- Arima(train, order=c(2,1,2), seasonal=c(2,1,2))
gdp_arima  <- forecast(model, h=7)
model
accuracy(gdp_arima, test)[, c(2,5,7,8)]
```

```r
model  <- Arima(train, order=c(3,1,0), seasonal=c(3,1,0))
gdp_arima  <- forecast(model, h=7)
model
accuracy(gdp_arima, test)[, c(2,5,7,8)]
```

```r
model  <- Arima(train, order=c(3,0,0), seasonal=c(3,0,0), include.drift = TRUE, include.constant
= TRUE)
gdp_arima  <- forecast(model, h=7)
model
accuracy(gdp_arima, test)[, c(2,5,7,8)]
```
