### eu

https://cran.r-project.org/web/packages/dtwclust/dtwclust.pdf

```r
df_eu  <-  read.csv(file="../data_hts/hts_eu_a10.csv", header=TRUE, sep=",", row.names =1 )
df_eu  <-  df_eu/1000
var_freq  <- 4    
var_start_year  <- 2000   
gdp_agg <- df_eu_sums <-  rowSums(df_eu)
gdp_agg  <- ts(gdp_agg, frequency = var_freq, start = var_start_year) 
```

```r
sc_df = apply(df_eu, 2, function(x) (x - mean(x))/sd(x) )
```

```r
ts.plot(sc_df)
```

```r
library('dtwclust')
```

```r
data  <-  sc_df
```

```r
pc <- tsclust(t(data), type = "partitional", k = 20L, 
              distance = "dtw_basic", centroid = "pam", 
              seed = 3247L, trace = TRUE,
              args = tsclust_args(dist = list(window.size = 5L)))
```

```r
plot(pc)
```

```r
interactive_clustering(t(data))
```

- type = Hierarchical
- distance measure = euclidean
- seed = 0
- n_clust = 25
- silhouette = 0.09


```r
plot(eu_clusted)
```

```r
eu_clusted$labels
```

```r
df  <-  cbind(as.data.frame(eu_clusted@cluster),  as.data.frame(eu_clusted@cluster))
```

```r
write.csv(df, file='index_list_eu.csv')
```
